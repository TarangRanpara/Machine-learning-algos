```python
import matplotlib.pyplot as plt
import numpy as np
from random import randint
```

```python
def line_plot():
    
    #line plot
    x = [1,2,3,4,5,6,7]
    y = [50,51,52,48,47,49,46]
    
    #plotting the data
    plt.plot(x,y,label='temp',color='cyan',linewidth='3')
    
    #attributes of plot
    plt.title('days & temp')
    plt.xlabel('days')
    plt.ylabel('temp')
    plt.legend()
    
    #shows the lines grid
    plt.grid()
    plt.show()
```

```python
line_plot()
```

```python
def trignometry_plot():
    
    #trignometry problem
    x = np.linspace(-np.pi, np.pi, 12)
    y = np.sin(x)
    z = np.cos(x)
    p = np.tan(x)
    
    #plotting the data
    plt.plot(x,y,label='sin',color='yellow',linewidth='3',marker='o',alpha=0.9)
    plt.plot(x,z,label='cos',color='red',linewidth = '3',marker='*',alpha=0.5)
    plt.plot(x,p,label='tan',color='blue',linewidth = '3',marker='+',alpha=0.9)
    
    #attributes of plot
    plt.title('Trignometry functions')
    plt.xlabel('angle')
    plt.ylabel('sin , cos & tan')
    plt.legend()
    
    #shows the lines grid
    plt.grid()
    plt.show()
```

```python
trignometry_plot()
```

```python
def bar_plot():
    
    #data
    company = ['GOOGL','AMZN','MSFT','FB']
    revanue = [90,136,89,27]
    profits = [40,2,34,12]
    
    xpos = np.arange(len(company))
    plt.bar(xpos,revanue,label='Revanue')
    plt.bar(xpos,profits,label='profits')
    
    #to replace name with numbers
    plt.xticks(xpos,company)
    
    #attributes of plot
    plt.xlabel('companies')
    plt.ylabel('revanue & profits')
    plt.title('company with their profits & revanue ')

    plt.legend()
    plt.show()
    
```

```python
bar_plot()
```

```python
def bar_plotH():
    
    #data
    company = ['GOOGL','AMZN','MSFT','FB']
    revanue = [90,136,89,27]
    profits = [40,2,34,12]
    
    #plotting Horizontally
    ypos = np.arange(len(company))
    plt.barh(ypos,revanue,label='Revanue')
    plt.barh(ypos,profits,label='profits')
    
    #to replace name with numbers
    plt.yticks(ypos,company)
    
    #attributes of plot
    plt.ylabel('companies')
    plt.xlabel('revanue & profits')
    plt.title('company with their profits & revanue ')

    plt.legend()
    plt.show()
```

```python
def hist_plot():
    
    #data
    blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
    blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
    
    #plotting
    plt.hist([blood_sugar_men,blood_sugar_women],bins = [40,70,100,130,160,180],rwidth=0.98,color = ['green','orange'],label=['men','women'])
    
    #attributes of plot
    plt.ylabel('frequency')
    plt.xlabel('sugar level')
    plt.title('people & their sugar level')
    
    plt.legend()
    plt.show()
```

```python
hist_plot()
```

```python
def scatter_plot():
    
    #data
    x = np.array([randint(1,10) for i in range(11)])
    y = np.array([randint(1,10) for i in range(11)])
    
    #plotting
    plt.scatter(x,y,label='data',marker='+',color='orange')
    plt.scatter(x + randint(1,10),y + randint(1,10),label='another data',marker='*',color='blue')
    
    #attributes of plot
    plt.ylabel('y')
    plt.xlabel('x')
    plt.title('scattered plot')
    
    plt.legend()
    plt.show()
```

```python
scatter_plot()
```

```python
def pie_plot():
    
    #pie chart
    expense_vals = [1400,600,300,410,250]
    expense_labels = ['home rent','food','phone/internet','car','others']
    
    plt.pie(expense_vals,labels = expense_labels,autopct='%0.2f%%',explode = [0.1,0,0,0,0])
    plt.title('__expenses__')
    plt.show()
```

```python
pie_plot()
```
