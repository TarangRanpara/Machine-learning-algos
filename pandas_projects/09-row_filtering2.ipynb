```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
#multiple conditions
#object.loc[rows,columns]
movies.loc[movies.star_rating>8,:].loc[movies.genre=='Crime'].head()
```

```python
movies.loc[movies.star_rating>8,:].loc[movies.genre=='Crime'].shape
```

```python
#good way to do this
#'and' -> '&'
#'or' -> '\'
movies[(movies.star_rating > 7) & (movies.genre == 'Crime') & (movies.duration > 150)].head()
```
